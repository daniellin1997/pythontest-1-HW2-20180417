## 3. 話不多說，直接進入正題！

### 首先，先設定好一個2x2的矩陣A

```python
A = np.array([[8,7],[5,1]])
```

這邊設個8751，藉以紀念Intel開發的8751 MCU 38週年（咦）

```python
A
```

### 接下來，第一個圖形：不通過原點的直線

這邊畫一條直線： $$ 2x+y=10 $$

```python
x = np.linspace(-10,10,200)
y = 10-2*x
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(x,y)
```

接著，讓A對它做線性轉換：

```python
xy = np.array([x,y])
Axy = A@xy

ax = plt.gca()
ax.set_aspect("equal")

plt.plot(Axy[0],Axy[1],'r')
```

它被拉長還被放倒了一點了呢！

### 第二個圖形：圓形（不過原點的，但原點是它的圓心）

```python
θ = np.linspace(0, 2*np.pi, 200)
r = 2
x = r*np.cos(θ)
y = r*np.sin(θ)
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(x,y)
```

接著，讓A對它做線性轉換：

```python
xy = np.array([x,y])
Axy = A@xy

ax = plt.gca()
ax.set_aspect("equal")

plt.plot(Axy[0],Axy[1],'r')
```

它被拉長...還被放倒了不只一點呢！

（下面加碼）把上面的那個傢伙再線轉個幾次

```python
A2xy = A@Axy
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(A2xy[0],A2xy[1],'g')
```

它變得更細ㄌ；再給它轉一次！

```python
A3xy = A@A2xy
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(A3xy[0],A3xy[1],'b')
```

細到我只想換下一個圖形了

### 第三個圖形：其他圖形

這次來點怪異的

```python
x = 3*(1-np.sin(10*θ))
y = np.cos(θ)
plt.plot(x,y)
```

就決定是這傢伙了！來讓它被A轉看看吧～
順便調換一下x跟y

```python
xy=np.c_[x,y]
xy=np.array([x,y])
Axy = A@xy
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(Axy[0],Axy[1],'r')
```

生出奇怪的東西了OAO

多轉兩次看看：

```python
A2xy = A@Axy
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(A2xy[0],A2xy[1],'y')
```

```python
A3xy = A@A2xy
ax = plt.gca()
ax.set_aspect("equal")
plt.plot(A3xy[0],A3xy[1],'g')
```

感覺好像一直線轉下去，各種圖形都會越接近直線呢！（在A是個2x2的矩陣時）
